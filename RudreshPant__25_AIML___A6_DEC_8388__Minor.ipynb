```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.histplot(df['age'], kde=True)
plt.title("Age Distribution of Football Players")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['overall'], kde=True)
plt.title("Overall Rating Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='age', y='overall', data=df)
plt.title("Age vs Overall Rating")
plt.show()
```

```python
cols = ['overall','potential','pace','shooting','passing','dribbling','defending','physic']

plt.figure(figsize=(10,6))
sns.heatmap(df[cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Player Attributes")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x='player_positions', y='overall', data=df)
plt.xticks(rotation=90)
plt.title("Position-wise Overall Rating")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(df['overall'])
plt.title("Outliers in Overall Rating")
plt.show()
```
